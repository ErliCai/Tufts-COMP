```python
# Problem 1 starts here

import networkx as nx
import matplotlib.pyplot as plt
q1 = nx.Graph()
Parks = ["Jumbo Studios","Jumbo Kingdom","World's Fair",
                   "Jumbo Golf Course","Trunk Water Park","Animal Planet Zoo"]
q1.add_nodes_from(Parks)
# color_map = ['blue','blue','blue','blue','blue','blue']

```

```python
# read from the JumboWorld.txt
f = open("JumboWorld.txt","r")

# The first line of the file is not needed
line = f.readline()
line = f.readline()

while(line):
    a = line.split('\t')
    q1.add_edge(a[0],a[2])
#   color_map.append('red')
    line = f.readline()
f.close()  

# Code to test degree methods in networkx
# for node in q1:
#     if node not in Parks:
#         print(q1.degree(node))
# print(q1.degree("Smith"))

# nx.draw(q1, node_color=color_map)
```

```python

import csv

csvFile = open("p1.csv", "w")
writer = csv.writer(csvFile)

header = [""] + Parks
writer.writerow(header)

for node in q1:
    if node not in Parks:
        entrence = [0,0,0,0,0,0]
        for adj in q1[node]:
            entrence[Parks.index(adj)] = 1
        writer.writerow([node]+entrence)
        
        

csvFile.close()

# print the number of edges for each park
# It should reflect the popularity of each park
for park in Parks:
    print("number of visitors to "+ park + ": ", end = " " ) 
    print( q1.degree(park) )
```

```python
#Problem 2 starts here

q2 = nx.Graph()
q2.add_nodes_from(Parks)
# Make nodes representing parks have color blue
color_map = ['blue','blue','blue','blue','blue','blue']

f = open("JumboWorld.txt","r")
line = f.readline()
line = f.readline()

family_list = {}


i = 1
while(line):
    a = line.split('\t')
    if a[0] not in family_list:
        family_list[a[0]] = {(a[1],a[2])}
    else:
        family_list[a[0]].add((a[1],a[2]))
    line = f.readline()
    i += 1

f.close()  

hotel_park = {}    

for family in family_list:
    for hotel, park in family_list[family]:
        if hotel not in hotel_park:
            hotel_park[hotel] = [0,0,0,0,0,0]
        hotel_park[hotel][Parks.index(park)] += 1
            
# print(hotel_park)

```

```python
csvFile = open("p2.csv", "w")
writer = csv.writer(csvFile)

header = [""] + Parks
writer.writerow(header)

for hotel in hotel_park:
    entry = [hotel] + hotel_park[hotel]
    writer.writerow(entry)
csvFile.close()

```

```python
# plot the graph
# Add edges to the graph and label the edge with their weight
edge_lable = {}
for hotel in hotel_park:
    color_map.append('red')
    for i in range(6):
        q2.add_edge(hotel,Parks[i],weight=hotel_park[hotel][i])
        edge_lable[(hotel,Parks[i])] = hotel_park[hotel][i]

plt.figure()
pos = nx.spring_layout(q2)

fig = plt.figure(figsize=(15, 15)) 
nx.draw(q2,pos, with_labels = True,node_color=color_map,)
nx.draw_networkx_edge_labels(q2,pos,edge_labels = edge_lable,font_color='red')
plt.axis('off')
plt.show()
```

```python
#question3 starts here

q3 = nx.Graph()

f = open("JumboWorld.txt","r")
line = f.readline()
line = f.readline()

i = 1
family_list2 = {}
last_family = None

while(line):
    a = line.split('\t')
    if a[0] != last_family:
        family_list2[a[0]] = {a[2]}
    else:
        family_list2[a[0]].add(a[2])
    line = f.readline()
    last_family = a[0]
    i += 1
# print(family_list2)
q3.add_nodes_from(family_list2)
for family1 in family_list2:
    for family2 in family_list2:
        if (family1 is not family2 and
        len(family_list2[family1].intersection(family_list2[family2])) >1):
            q3.add_edge(family1,family2)

print(q3.degree('Smith'))
print(q3.degree('Robinson'))
# print(header)
# print(q3.has_edge('Smith','Brown'))
```

```python
csvFile = open("p3.csv", "w")
writer = csv.writer(csvFile)

header = [family for family in family_list2]
header = [''] + header
writer.writerow(header)

for f1 in family_list2:
    ans = [f1]
    for f2 in family_list2:
        if f1 == f2 or q3.has_edge(f1,f2):
            ans.append(1)
        else:
            ans.append(0)
            
    writer.writerow(ans)


    
csvFile.close()


```

```python
#question4 starts here

q4 = nx.Graph()

f = open("JumboWorld.txt","r")
line = f.readline()
line = f.readline()

family_list3 = {}
last_family = None

i = 1
while(line):
    a = line.split('\t')
    if a[0] != last_family:
        family_list3[a[0]] = {(a[2],a[3])}
    else:
        family_list3[a[0]].add((a[2],a[3]))
    line = f.readline()
    last_family = a[0]
    i += 1
# print(family_list2)
q4.add_nodes_from(family_list3)
for family1 in family_list3:
    for family2 in family_list3:
        if (family1 is not family2 and
        len(family_list3[family1].intersection(family_list3[family2])) >1):
            q4.add_edge(family1,family2)
            
# Compare the results from these two lines from what we get from question 3,
# we see that some nodes have more edges and some has less
# This is because Robinson go to the same park 3 times and it won't contribute to any edges in q3.
# while in the case of Smith family, they go to the same park as many other families(so many edges in q3)
# but they are in different days, so it won't form edges in q4
print(q4.degree('Smith'))
print(q4.degree('Robinson'))


            
csvFile = open("p4.csv", "w")
writer = csv.writer(csvFile)

header = [family for family in family_list3]
header = [''] + header
writer.writerow(header)

for f1 in family_list3:
    ans = [f1]
    for f2 in family_list3:
        if f1 == f2 or q4.has_edge(f1,f2):
            ans.append(1)
        else:
            ans.append(0)
            
    writer.writerow(ans)


    
csvFile.close()


```
