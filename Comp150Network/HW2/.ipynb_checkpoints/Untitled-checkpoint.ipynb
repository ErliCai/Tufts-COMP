```python
# Question 1

import networkx as nx
import matplotlib.pyplot as plt
file = open("karate/zachary.dat")
f = file.readlines()

karate = nx.Graph()
karate.add_nodes_from(list(range(34)))

for i in range(34):
    line = f[i+7].split()
    for j in range(34):
        if line[j] == '1':
            karate.add_edge(i,j)
fig = plt.figure(figsize=(15, 15)) 

color_map = ['blue']*34
color_map[0] = 'green'

nx.draw(karate,node_size = 200,edge_color='r',node_color=color_map)


print(nx.diameter(karate))
# The diameter of this graph is 5 
# and since the result is less than infinity we know that the graph is connected

print(max([karate.degree(x) for x in karate]))
# Maximum degree is 17

# print(len(karate))
print(nx.average_clustering(karate))

# print(nx.shortest_path(karate,0,33))


```

```python
shortest_length = []
for i in range(34):
    for j in range(i+1,34):
        shortest_length.append(len(nx.shortest_path(karate,i,j))-1)
# print(shortest_length)
plt.hist(shortest_length,bins=5,range=(0.5,5.5))



```

```python
Degree = [karate.degree(node) for node in karate]
plt.hist(Degree,bins=17,range=(0.5,17.5))

```

```python
betweeness_map = nx.betweenness_centrality(karate)
n = 0
maximum = 0
for i in range(34):
    if betweeness_map[i] > maximum:
        maximum = betweeness_map[i]
        n = i
print(n)
betweeness = list(betweeness_map.values())
plt.hist(betweeness,bins=30)
```

```python
clustering_coefficient = nx.clustering(karate) 
c = list(clustering_coefficient.values())
plt.hist(c,bins=20)
```

```python
# Question 3

import networkx as nx
import matplotlib.pyplot as plt
file = open("prison/prison.dat")
f = file.readlines()

prison = nx.Graph()
prison.add_nodes_from(list(range(34)))

for i in range(67):
    line1 = f[2*i+4].split()
    line2 = f[2*i+5].split()
    line = line1 + line2
    for j in range(67):
        if line[j] == '1':
            prison.add_edge(i,j)
fig = plt.figure(figsize=(15, 15)) 

color_map = ['blue']*100
color_map[15] = 'green'

nx.draw(prison,node_size = 200,edge_color='r',node_color=color_map)


print(nx.diameter(prison))
# The diameter of this graph is 7 
# and since the result is less than infinity we know that the graph is connected

print(max([prison.degree(x) for x in prison]))
# Maximum degree is 11

# print(len(prison))
print(nx.average_clustering(prison))

# print(nx.shortest_path(prison,0,33))

```

```python
shortest_length = []
for i in range(67):
    for j in range(i+1,67):
        shortest_length.append(len(nx.shortest_path(prison,i,j))-1)
# print(shortest_length)
plt.hist(shortest_length,bins=7,range=(0.5,7.5))

```
