```python
#Question 1

import networkx as nx
import matplotlib.pyplot as plt

p1 = nx.Graph()

caida_rel = open('caida/as_rel.txt')
l = caida_rel.readline()
line = l.split()

while l:
    line = l.split()
    if line[2] == '0':
        p1.add_edge(line[0], line[1])
    l = caida_rel.readline()
    
    
caida_rel.close()

fig = plt.figure(figsize=(30, 30)) 
nx.draw(p1,node_size = 60, edge_color='r')
```

```python
import random
nodes = list(p1.nodes)

FourHundred_nodes = []
while len(FourHundred_nodes) < 400:
    a = random.choice(nodes)
    if a not in FourHundred_nodes:
        FourHundred_nodes.append(a)
        
Other_nodes = []
for node in nodes:
    if node not in FourHundred_nodes:
        Other_nodes.append(node)

shortest_path_length = []
for i in FourHundred_nodes:
    for j in Other_nodes:
        s = nx.shortest_path_length(p1, i, j)
        shortest_path_length.append(s)

plt.hist(shortest_path_length,bins=10, range=(0.5,10.5))
        
```

```python
#Question 2
p2 = nx.Graph()
p2 = nx.k_core(p1,2)
print(len(p2.nodes))
fig = plt.figure(figsize=(30, 30)) 
nx.draw(p2,node_size = 60, edge_color='r')

```

```python
print(nx.diameter(p2))
print(nx.average_shortest_path_length(p2))
    
```

```python
#Question 3
print(len(p2.nodes))
print(len(p2.edges))

def remove_n_nodes(percentage):
    p3 = p2.copy()
    n = int(len(p2.nodes) * percentage /100)
    for i in range(n):
        p3.remove_node(random.choice(list(p3.nodes)))
    largetest_connected_component = sorted(nx.connected_components(p3), key=len, reverse=True)[0]
    number_of_nodes = len(largetest_connected_component)
    ncp = nx.number_connected_components(p3)
    p3 = p3.subgraph(largetest_connected_component)
    asp = nx.average_shortest_path_length(p3)
    diameter = nx.diameter(p3)
    print(percentage,'% nodes removal:',ncp ,'connected components,','average shortest path length is ',asp, ', diameter is',diameter,)
    
def remove_n_edges(percentage):
    p3 = p2.copy()
    n = int(len(p2.edges) * percentage /100)
    for i in range(n):
        e = random.choice(list(p3.edges))
        p3.remove_edge(e[0], e[1])
    largetest_connected_component = sorted(nx.connected_components(p3), key=len, reverse=True)[0]
    number_of_nodes = len(largetest_connected_component)
    ncp = nx.number_connected_components(p3)
    p3 = p3.subgraph(largetest_connected_component)
    asp = nx.average_shortest_path_length(p3)
    diameter = nx.diameter(p3)
    print(percentage,'% edges removal:',ncp ,'connected components,','average shortest path length is ',asp, ', diameter is',diameter,)

for i in range(20):    
    remove_n_nodes(5)
    remove_n_nodes(10)
    remove_n_nodes(15)
    remove_n_edges(5)
    remove_n_edges(10)
    remove_n_edges(15)
```

```python
#Question 4

nodes = list(p2.nodes)

size_of_minimum_edge_cut = []

for i in range(40):
    n1 = random.choice(nodes)
    n2 = n1
    while n2 == n1:
        n2 = random.choice(nodes)
    s = nx.minimum_edge_cut(p2,n1,n2)
    size_of_minimum_edge_cut.append(len(s))
    print('The min set of edges to be removed between ',n1, ' and ',n2,': ', s )
n = max(size_of_minimum_edge_cut)
plt.hist(size_of_minimum_edge_cut, bins = n, range = (0.5,n+0.5))    
    
        
nx.minimum_edge_cut(p2,'6140','8190')
```
