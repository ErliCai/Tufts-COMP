```python
import networkx as nx
import matplotlib.pyplot as plt

p1 = nx.Graph()

caida_rel = open('caida/as_rel.txt')
l = caida_rel.readline()
line = l.split()

while l:
    line = l.split()
    if line[2] == '0':
        p1.add_edge(line[0], line[1])
    l = caida_rel.readline()
    
    
caida_rel.close()

fig = plt.figure(figsize=(30, 30)) 
nx.draw(p1,node_size = 60, edge_color='r')
```

```python
import random
nodes = list(p1.nodes)

FourHundred_nodes = []
while len(FourHundred_nodes) < 400:
    a = random.choice(nodes)
    if a not in FourHundred_nodes:
        FourHundred_nodes.append(a)
        
Other_nodes = []
for node in nodes:
    if node not in FourHundred_nodes:
        Other_nodes.append(node)

shortest_path_length = []
for i in FourHundred_nodes:
    for j in Other_nodes:
        s = nx.shortest_path_length(p1, i, j)
        shortest_path_length.append(s)

plt.hist(shortest_path_length,bins=10, range=(0.5,10.5))
        
```

```python
p2 = nx.Graph()
p2 = nx.k_core(p1,2)
print(len(p2.nodes))
fig = plt.figure(figsize=(30, 30)) 
nx.draw(p2,node_size = 60, edge_color='r')

```

```python
print(nx.diameter(p2))
print(nx.average_shortest_path_length(p2))
    
```
